## Ejemplo 1

```python
import random
import math
import matplotlib.pyplot as plt
def generar_punto():
    while True:
        x=random.uniform(-1, 1)
        y=random.uniform(-1, 1)
        if(math.sqrt(x**2 + y**2)<=1):
            return x, y
cantidad_puntos=1000
puntos=[generar_punto() for _ in range(cantidad_puntos)]
x_puntos=[punto[0] for punto in puntos]
y_puntos=[punto[1] for punto in puntos]
fig, ax=plt.subplots()
circulo=plt.Circle((0, 0), 1, color='b', fill=False)
ax.add_artist(circulo)
ax.scatter(x_puntos, y_puntos, color='r', s=1)
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal', 'box')
ax.grid(True)
plt.show()
```

## Ejemplo 2

```python
import matplotlib.pyplot as plt
import random
import math
V1=(0, 0)
V2=(1, 0)
V3=(0.5, math.sqrt(3)/2)
vertices=[V1, V2, V3]
P=(0.5, 0.5)
puntos=[P]
iteraciones=10000
for _ in range(iteraciones):
    vertice=random.choice(vertices)
    P=((P[0]+vertice[0])/2, (P[1]+vertice[1])/2)
    puntos.append(P)
with open("triangle_puntos.txt", "w") as file:
    for point in puntos:
        file.write(f"{point[0]} {point[1]}\n")
x_coords, y_coords=zip(*puntos)
plt.scatter(x_coords, y_coords, s=0.1, color="blue")
plt.title("Puntos generados en el triángulo equilátero")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
